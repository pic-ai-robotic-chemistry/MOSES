```python
import os
import pymupdf4llm as PyMuPDF4LLM
from langchain.text_splitter import MarkdownHeaderTextSplitter



def process_pdf_folder(folder_path):
    """处理文件夹中的所有PDF文件,创建并更新对应的Markdown文件"""
    # 获取pdf文件列表
    pdf_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.lower().endswith('.pdf')]
    
    # 处理每个pdf文件
    for pdf_path in pdf_files:
        pdf_name = os.path.splitext(os.path.basename(pdf_path))[0]
        md_path = os.path.join(folder_path, f"{pdf_name}.md")
        
        # 转换pdf为markdown
        markdown_text = PyMuPDF4LLM.to_markdown(pdf_path)
        
        # 写入markdown文件(无论是否存在都会覆盖)
        with open(md_path, "w", encoding="utf-8") as f:
            f.write(markdown_text)
            
        print(f"已处理: {pdf_name}.pdf -> {pdf_name}.md")

def check_pdf_markdown_status(folder_path):
    """检查PDF文件的Markdown对应情况及文件大小"""
    # 获取所有pdf和md文件
    pdf_files = [f for f in os.listdir(folder_path) if f.lower().endswith('.pdf')]
    md_files = [f for f in os.listdir(folder_path) if f.lower().endswith('.md')]
    
    # 检查每个pdf文件
    for pdf_file in pdf_files:
        pdf_name = os.path.splitext(pdf_file)[0]
        md_file = f"{pdf_name}.md"
        
        if md_file in md_files:
            # 获取md文件字符数
            md_path = os.path.join(folder_path, md_file)
            with open(md_path, 'r', encoding='utf-8') as f:
                char_count = len(f.read())
            print(f"{pdf_file} -> {md_file} (字符数: {char_count})")
        else:
            print(f"{pdf_file} -> 未找到对应的Markdown文件")
```

```python
def split_markdown_to_chunks(folder_path):
    """读取文件夹中的所有markdown文件并分割为chunks"""
    # 获取所有md文件
    md_files = [f for f in os.listdir(folder_path) if f.lower().endswith('.md')]
    
    # 定义headers_to_split_on
    headers_to_split_on = [
        # ("#", "Header 1"),
        # ("##", "Header 2"), 
        ("######", "Header 6"),
    ]
    
    # 创建splitter实例
    markdown_splitter = MarkdownHeaderTextSplitter(
        headers_to_split_on=headers_to_split_on, strip_headers=False
    )
    
    all_chunks = []
    # 处理每个markdown文件
    for md_file in md_files:
        md_path = os.path.join(folder_path, md_file)
        # 读取markdown文件
        with open(md_path, 'r', encoding='utf-8') as f:
            markdown_text = f.read()
        
        # 分割文本
        chunks = markdown_splitter.split_text(markdown_text)
        all_chunks.extend(chunks)
        
    return all_chunks
```

```python
from src.agents.builder_team.concept_extractor import llm, generate_prompt

def test_individual_extraction(content, vanilla=False):
    """测试从文献文本中提取个体实例的基本流程"""
    
    # 准备测试用的文献文本列表
    test_texts = content
    # 获取系统提示词
    vanilla_prompt = r"""# Ontology Classification in Chemistry

## Task Overview

In the field of chemistry, we aim to classify concepts as either a **"class"** or an **"individual"** within an ontological framework.

## Definitions

- **Class:** A class represents a broad category or group that can include multiple entities or concepts. If a concept A can be linked to a set of other concepts or entities {B}, and it can be stated that {B} are examples of A or belong to A, then A is a class.

- **Individual:** An individual is a distinct, singular entity or concept that does not include other entities. It is unique and cannot be subdivided into further examples or categories.

## Task Instructions

1. **Analyze the Text:** Thoroughly read the provided text to understand the context.
2. **Extract Concepts:** Identify and extract all relevant concepts and entities from the text.
3. **Classify Each Concept:**
   - If a concept can be associated with multiple examples or entities, classify it as a **"class."**
   - If the concept is unique and specific, classify it as an **"individual."**
4. **Justify Your Classification:** Provide a concise explanation for each classification to support your decision.

## Example

- **Text:** "Water is a compound that consists of hydrogen and oxygen. Elements like hydrogen and oxygen are fundamental to chemistry."
  - **Compound** is a **class** because it can encompass various examples such as water, carbon dioxide, etc.
  - **Water** is an **individual** because it is a specific instance of a compound.
  - **Element** is a **class** because it includes examples like hydrogen, oxygen, etc.
  - **Hydrogen** and **Oxygen** are **individuals** because they are specific elements.
  
For each identified concept, output only:
Name: [Specific concept name/identifier]
Classification: [Class or Individual]
Justification: [Explanation for the classification]

If the concept has abbreviations, format its Name [Full name]([Abbreviation]).

Apply this structured approach to analyze the text below and accurately classify the concepts.
"""
    # 根据vanilla参数选择使用的提示词
    prompt = vanilla_prompt if vanilla else generate_prompt()
    
    # 对每段文本进行处理
    for text in test_texts:
        # 调用大模型进行提取
        response = llm.invoke(f"{prompt}\n\nInput text:\n{str(text)}")
        print(f"\n提取结果:\n{response.content}")
```
